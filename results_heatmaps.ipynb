# Heatmaps of Results

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# load data
csv_path_q1 = '/Users/saraabesova/Documents/Q1_final results.csv'
df_q1 = pd.read_csv(csv_path_q1)

# print(df_q1.head())
plt.rcParams['font.family'] = 'serif'  # font set to serif

colors = ["#e6f2f2", "#b2d8d8", "#80b2b2"]
cmap = mcolors.LinearSegmentedColormap.from_list("custom_light_teal", colors, N=100)

# initialise the heatmap
plt.figure(figsize=(6, 5))
ax = sns.heatmap(df_q1.set_index(df_q1.columns[0]), annot=True, cmap=cmap)  

# get the position of the dotten line -- down the middle of the columns
middle_column_position = df_q1.shape[1] // 2

# place the dotted line
ax.axvline(middle_column_position, color='black', linestyle='--', linewidth=2)

plt.xticks(color='grey') 
plt.yticks(color='grey')
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha="right")  


ax.set_ylabel('Relations', fontsize=11)
ax.set_xlabel('Related Objects                  Unrelated Objects', font='serif', fontsize=11, color='black')

plt.title('Spatial Accuracy (Q1)', font='serif', fontsize=15, color='black', pad=16)
plt.tight_layout()
plt.savefig('/Users/saraabesova/Documents/heatmap_Q1.png', format='png', dpi=300)
plt.show()
```
